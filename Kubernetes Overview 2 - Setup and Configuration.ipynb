# Setup and Configuration
We're gonna first start with installation.

## Installation
In order for this to work, we need to have 4 servers running - 3 minions and 1 master - so we'll set that up with docker.

### Setting up docker machines

```bash
# Master
docker run -dit --name=master -p 8000:80 centos:latest /bin/bash
# Minion1
docker run -dit --name=minion1 -p 8001:80 centos:latest /bin/bash
# Minion2
docker run -dit --name=minion2 -p 8002:80 centos:latest /bin/bash
# Minion3
docker run -dit --name=minion3 -p 8003:80 centos:latest /bin/bash
```

## Install NTPD 
We need to makes sure that ntpd is running in all instances

We want to install npd because we need to know what's happening, and that the servers are reporting.

```bash
docker exec master yum install -y ntp
docker exec minion1 yum install -y ntp
docker exec minion2 yum install -y ntp
docker exec minion3 yum install -y ntp
```

### Now let's make sure it's running

```bash
docker exec master bash -c "systemctl enable ntpd && systemctl start ntpd"
docker exec minion1 bash -c "systemctl enable ntpd && systemctl start ntpd"
docker exec minion2 bash -c "systemctl enable ntpd && systemctl start ntpd"
docker exec minion3 bash -c "systemctl enable ntpd && systemctl start ntpd"
```

### Let's check that it has been installed

```bash
docker exec master bash -c "systemctl status ntpd"
```
